```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# degree --- 0 to 360
```

```python
angles=np.arange(0,361,15)
```

```python
print(angles)
```

```python
rad=np.deg2rad(angles)
```

```python
print(rad)
```

```python
sin=np.sin(rad)
```

```python
cos=np.cos(rad)
```

```python
print(sin)
```

```python
plt.figure(figsize=(12,5))
plt.plot(angles,sin,ls=':',lw=4,color='red',label='sin')
plt.plot(angles,cos,ls='--',lw=3,color='g',label='cos')
plt.legend()
plt.xlabel('Angles',fontsize=15,color='green')
plt.ylabel('Values',fontsize=15,color='g')
plt.title('Sin Vs Cos',fontsize=18,color='g')
plt.grid()
plt.show()
```

```python
data=np.linspace(-8,8,150)
```

```python
print(data)
```

```python
sin=np.sin(data)
cos=np.cos(data)
```

```python
plt.figure(figsize=(10,5))
plt.plot(data,sin,color='black',ls='--',lw=3,label='sin')
plt.plot(data,cos,color='red',ls=':',lw=3,label='cos')
plt.legend()
plt.grid()
plt.show()
```

```python
fig=plt.figure(facecolor='red',figsize=(9,4))
ax=fig.add_axes([0,0,1,1],facecolor='black')
x=np.linspace(-5,5,10)
ax.plot(x,x**2,color='red',ls='--',lw=3,label='square',marker='o')
ax.plot(x,x**3,color='white',ls=':',lw=3,label='qube',marker='s')
ax.set_xticklabels(["A",'B','C','D','E','F'],fontsize=12,color='white',fontweight='bold',rotation=45,fontfamily='monospace')
ax.legend()
ax.set_xlabel('X-Values',color='white',fontweight='bold',fontfamily='monospace',fontsize=15)
ax.set_title('Line plot with facecolor',color='white',fontweight='bold',fontfamily='monospace',fontsize=15)
ax.grid()
plt.show()
```
