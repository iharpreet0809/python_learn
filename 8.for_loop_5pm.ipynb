```python
p='fyufyu%45644343234$@#$@4cvcfdyt'
```

```python
alpha=''
digit=''
special=''
for i in p:
    if i.isalpha():
        alpha=alpha+i
    elif i.isdigit():
        digit=digit+i
    else:
        special=special+i
print('alphabets: ',alpha)
print('digits: ',digit)
print('special char: ',special)
```
